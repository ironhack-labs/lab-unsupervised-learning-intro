```python
# Importing libraries
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing data
data = pd.read_csv(r"C:\Users\joaoa\Desktop\Ironhack\Labs\lab-unsupervised-learning-intro\dataset_spotify.csv")
data = data.iloc[:,1:]
data
```

```python
# Creating a new dataframe with numerical values
df_num = data.select_dtypes(include=np.number)
df_num
```

```python
# Standardizing numerical columns
from sklearn.preprocessing import StandardScaler
stand = StandardScaler().fit_transform(df_num)
df_stand = pd.DataFrame(stand, columns = df_num.columns)
df_stand
```

# Clustering

```python
# Elbow
from sklearn.cluster import KMeans
K = range(1, 11)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=0)
    kmeans.fit(df_stand)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbowethod showing the optimal k')
plt.show()
```

```python
# Silhouette
from sklearn.metrics import silhouette_score
K = range(2, 11)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state = 0)
    kmeans.fit(df_stand)
    silhouette.append(silhouette_score(df_stand, kmeans.predict(df_stand)))


plt.figure(figsize=(16, 8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K) + 1, 1.0))
plt.title('Silhouette method showing the optimal k')
plt.show()
```

```python
# Creating 2 clusters
kmeans = KMeans(n_clusters = 2, random_state = 0)
kmeans.fit(df_stand)
clusters = kmeans.predict(df_stand)
data["cluster"] = clusters
data
```
