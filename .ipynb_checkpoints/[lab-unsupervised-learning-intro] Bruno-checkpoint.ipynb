```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
m_data = pd.read_csv(r"C:\Users\btdjf\Desktop\Ironhack 2\Lab Unsupervised Learning Intro\lab-unsupervised-learning-intro\music_data.csv")
m_data
```

```python
kaggle = pd.read_csv(r"C:\Users\btdjf\Desktop\Ironhack 2\Lab Unsupervised Learning Intro\lab-unsupervised-learning-intro\dataset.csv")
kaggle
```

```python
audio_features = kaggle[[ 'danceability', 'energy', 'key', 'loudness', 'mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo']]
audio_features
```

```python
scaler = StandardScaler()
scaled_audio_features = scaler.fit_transform(audio_features)
```

```python
df = pd.DataFrame(scaled_audio_features, columns=audio_features.columns)
df
```

```python
kmeans = KMeans(n_clusters=8, random_state=1234)
kmeans.fit(scaled_audio_features)
```

```python
cluster = kmeans.predict(scaled_audio_features)
cluster
```

```python
pd.Series(cluster).value_counts().sort_index()
```

```python
new_df = pd.DataFrame(df)
new_df["cluster"] = cluster
new_df
```

```python
kmeans.inertia_
```

```python
kmeans = KMeans(n_clusters=8,
                init="random",
                n_init=5, 
                max_iter=2,
                tol=0,
                algorithm="full",
                random_state=1234)
kmeans.fit(scaled_audio_features)
print(kmeans.inertia_)
```

```python
K = range(2, 20)
inertia = []
for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=0)
    kmeans.fit(scaled_audio_features)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
K = range(2, 20)
silhouette = []
for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=0)
    kmeans.fit(scaled_audio_features)
    silhouette.append(silhouette_score(scaled_audio_features, kmeans.predict(scaled_audio_features)))
plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```
