```python
# Clean the data

# Drop the column i do not need except song name and artist
cols_to_drop = ['track_id', 'album_name', 'type', 'uri', 'track_href','analysis_url','duration_(ms)'] 

data = playlist.drop(columns=cols_to_drop)
data
```

```python
# Convert 'release_date' to datetime
data['release_date'] = pd.to_datetime(data['release_date'], errors = 'coerce', format='%Y-%m-%d')

# Create columns for year and month
data['year'] = data['release_date'].dt.year
data['month'] = data['release_date'].dt.month

# Drop 'release_date' and timestamp null values
data.drop('release_date', axis=1, inplace=True)
data = data.dropna()
```

```python
# Display data
analyze_data(data)
```

```python
X_data = data.copy()
```

```python
# Standardize 
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Assuming 'df' is your DataFrame

# Define the categorical columns
categorical_cols = ['key', 'mode', 'time_signature','song_name','artists']

# Select the numerical columns (excluding categorical ones)
numerical_cols = X_data.columns.difference(categorical_cols)

# Initialize the StandardScaler
scaler = StandardScaler()

# Scale the numerical columns
X_data[numerical_cols] = scaler.fit_transform(X_data[numerical_cols])

X_data
```

```python
# Encode categorical

# 'key'-> The key the track is in. Integers map to pitches using standard Pitch Class notation.

# 'time_signature' ->  The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure).

# One-hot encode 'key' feature
key_dummies = pd.get_dummies(X_data['key'], prefix='key')

# One-hot encode 'time_signature' feature
time_signature_dummies = pd.get_dummies(X_data['time_signature'], prefix='time_signature')

# Drop the original 'key' and 'time_signature' columns
X_data = X_data.drop(['key', 'time_signature'], axis=1)
```

```python

# Correct approach to concatenate 'data' with 'key_dummies' and 'time_signature_dummies'
X_data = pd.concat([X_data, key_dummies, time_signature_dummies], axis=1)
X_data
```

## Clustering model

### K means

```python
X_data = X_data.drop(columns = ['song_name','artists'], axis = 1)
```

```python
## Choosing the right number of Kmeans

# Elbow method
from sklearn.cluster import KMeans
import numpy as npxsw35

K = range(2, 30)
inertia = []
for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=0)
    kmeans.fit(X_data)
    inertia.append(kmeans.inertia_)
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
# Silhouette score
from sklearn.metrics import silhouette_score

K = range(2, 30)
silhouette = []
for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=0)
    kmeans.fit(X_data)
    silhouette.append(silhouette_score(X_data, kmeans.predict(X_data)))
plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Silhouette Method showing the optimal k')
```

```python
# two elbows here are interesting to consider
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

# Assuming X_data is your dataset
model = KMeans(random_state=0)
visualizer = KElbowVisualizer(model, k=(2,20))

visualizer.fit(X_data)  # Fit the data to the visualizer
visualizer.show()       # Finalize and render the figure
```

```python
# Compute the model 
kmeans = KMeans(n_clusters = 10, random_state = 0)

kmeans.fit(X_data)

```

```python
# Predict clusters
cluster = kmeans.predict(X_data)

# Display number of clusters
pd.Series(cluster).value_counts().sort_index()
```
