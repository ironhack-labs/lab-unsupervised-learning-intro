```python
# Getting just the numerical values 
numerical = spotify_df.select_dtypes(include='number')
```

```python
numerical = numerical.drop('Unnamed: 0', axis=1) 
```

```python
numerical.describe()
```

```python
from sklearn.preprocessing import StandardScaler

X_prep = StandardScaler().fit_transform(numerical)
X_prep 
```

```python
X_df = pd.DataFrame(X_prep, columns = numerical.columns)
X_df.describe()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 5, random_state =42)
kmeans.fit(X_df)
```

```python
clusters = kmeans.predict(X_prep)
clusters
```
