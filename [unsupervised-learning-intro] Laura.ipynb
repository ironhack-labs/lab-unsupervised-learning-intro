Using only song_features to create model

```python
X = songs[['danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness',
           'liveness','valence','tempo']]
X
```

Standarisation

```python
from sklearn.preprocessing import StandardScaler

X_prep = StandardScaler().fit_transform(X)

# make it a df again to see the .decripbe()
X_df = pd.DataFrame(X_prep, columns = X.columns)
X_df.describe()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 8, random_state = 0)
kmeans.fit(X_prep)
```

```python
kmeans.inertia_
```

```python
clusters = kmeans.predict(X_prep)
clusters
```

```python
X_df_upd = pd.DataFrame(X_df)
X_df_upd["cluster"] = clusters
X_df_upd
```

```python
# Evaluation Matrix
K = range(2, 20)
inertia = []
for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=0)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
kmeans = KMeans(n_clusters = 14, random_state = 0)
kmeans.fit(X_prep)

kmeans.inertia_
```

```python
from sklearn.metrics import silhouette_score
K = range(2, 20)
silhouette = []
for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=0)
    kmeans.fit(X_prep)
    silhouette.append(silhouette_score(X_prep, kmeans.predict(X_prep)))
plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 14, random_state = 0)
kmeans.fit(X_prep)
kmeans.inertia_
```

```python
labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

# Plotting
plt.figure(figsize=(20, 10))
colors = plt.cm.rainbow(np.linspace(0, 1, len(set(labels))))

for i, color in zip(range(len(cluster_centers)), colors):
    # Plot data points belonging to the current cluster
    plt.scatter(X_prep[labels == i, 0], X_prep[labels == i, 1], color=color, label=f'Cluster {i}')
    # Plot the cluster center
    plt.scatter(cluster_centers[i][0], cluster_centers[i][1], color='black', marker='x', s=100)

plt.title('Clusters formed by KMeans')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
